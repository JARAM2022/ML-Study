```python
import random
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return 5*x - 1000

x = []
for i in range(100):
    x.append(random.uniform(-100,100))
    
y = []
for i in range(100):
    error = random.uniform(-100,100)
    y.append(f(x[i]) + error)
    
```

```python
plt.scatter(x, y)
plt.show()
```

```python
epochs = 30000
lr = 0.0002
```

```python
class lin_model():
    def __init__(self):
        self.w = random.random()
        self.b = random.random()
        
    def predict(self, x):
        return self.w*x + self.b
```

```python
def train(X, Y, model):
    dw = 0.0
    db = 0.0
    cost = 0.0
    m = len(X)
    for x, y in zip(X, Y):
        pred = model.predict(x)
        cost += (pred-y)**2
        dw += 2*(pred-y)*x
        db += 2*(pred-y)
    cost /= m
    model.w -= lr*(dw/m)
    model.b -= lr*(db/m)
    return cost
    
```

```python
model = lin_model()
```

```python
for epoch in range(epochs):
    cost = train(x, y, model)
    if epoch%1000 == 0:
        print(cost)
print(str(model.w) + ' ' + str(model.b))
print('train_finished')
    
```

```python
pred1 = model.predict(-100)
pred2 = model.predict(100)
plt.scatter(x, y)
plt.plot([-100,100], [pred1, pred2], color = "blue", linewidth = 3)
plt.show()
```
